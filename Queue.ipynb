# Queue Model

```python
from pythonds.basic.queue import Queue
from random import randint 
from tqdm import tqdm
from math import ceil
import matplotlib.pyplot as plt

myqueue = Queue()
probablity_of_joining_cue = 0.3
no_of_cycles_before_item_leaves_queue = 1

time = []
no_of_items_in_a_queue = []
for i in tqdm(range(1,1000)):
    a = randint(0,ceil(1/probablity_of_joining_cue))
    if a == 0: 
        myqueue.enqueue(1)
        time.append(i)
    if len(time) > 0:
        if time[-1] + no_of_cycles_before_item_leaves_queue == i:
            myqueue.dequeue()
            time.pop(-1)
    no_of_items_in_a_queue.append(myqueue.size())

print(no_of_items_in_a_queue[:100])
plt.plot(no_of_items_in_a_queue[:100])
plt.title("Queue Behaviour between 1:100")
plt.show()
plt.plot(no_of_items_in_a_queue)
plt.title("Queue Behaviour between 1:1000")
plt.show()
```
